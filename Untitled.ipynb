```python
def compute_band_matrix(dataset=None, bm_size=1000, num_sensitive=1):
    """
        Function for computing the band matrix,
        Also choses sensitive items randomly 
        
        Params:
        .......
        dataset : DataFrame
            Original data (binary format transaction data)
            
        bm_size : int
            Size of the square band matrix
            
        num_sensitive: int 
            Number of sensitive items to be chosen randomly
        
    """
    if dataset is not None and len(dataset) >= bm_size and len(dataset.columns) >= bm_size:
        
        print(f'{"-"*20}\nFIRST CASE\n{"-"*20}')
        # list item names
        
        items = list(dataset.columns)
        # dataset = dataset.reset_index()
        # dataset.drop('index',axis=1,inplace=True)
        # shuffle the rows and columns of the dataset w.r.t bm_size
        np.random.seed(seed=13)
        random_column = np.random.permutation(dataset.shape[1])[:bm_size]
        random_row = np.random.permutation(dataset.shape[0])[:bm_size]
        
        print(f'items: {items}')
        # new order of items after shuffling
        items_reordered = [items[i] for i in random_column]
        
        print(f'\n\nRANDOM COLS: {random_column}\n\n')
        print(f'\n\nDataframe: {df.head()}\n\n')
        
        
        # cut selected square piece from the dataset
        df_square = dataset.iloc[random_row,random_column]
    

        f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
        
        # check if sensitive items are null
        sensitive_list = df_square.columns[-num_sensitive:]
        print(f'\n\nSensitive Items: {sensitive_list}\n')
        
        # plot the sparse square matrix cut from the original dataset
        ax1.spy(df_square, marker='.', markersize='1')
        # ax1.show()
        
        # apply Reverse Cuthill-McKee (RCM) algorithm
        sparse = csr_matrix(df_square)
        order = reverse_cuthill_mckee(sparse)

        # reorder the items a second time after the RCM algorithm
        items_final = [items_reordered[i] for i in order]
        column_reordered = [df_square.columns[i] for i in order]
        items_final = dict(zip(column_reordered,items_final))

        # banded matrix
        df_square_band = df_square.iloc[order][column_reordered]
        
        # plot banded matrix 
        ax2.spy(df_square_band, marker='.', markersize='1')
        #ax2.show()
        plt.show()
        # initial matrix band
        [i, j] = np.where(df_square == 1)
        bw = max(i-j) + 1
        print("Bandwidth first RCM", bw)

        # matrix band after RCM
        [i, j] = np.where(df_square_band == 1)
        bw = max(i-j) + 1
        print("Bandwidth after RCM", bw)
        return df_square_band, items_final, sensitive_list
    
    elif dataset is not None and len(dataset) >= bm_size:
        
        '''The case when the number of features (columns) are within the selected 
           dimension size
        '''
        print(f'{"-"*20}\nSECOND CASE\n{"-"*20}')
        
        # check the number of rows
        # then square them according to the given dimension size
        # cut the dataset 
        items = dataset.columns
        
        random_row = np.random.permutation(dataset.shape[0])
        logger('random row', random_row)
        dataset = dataset.iloc[random_row][:bm_size]
        dataset = dataset.reset_index()
        dataset.drop('index',axis=1,inplace=True)
        
        columns = dataset.columns
        
        # fill the gap between the features and wanted dimension size with zeros
        zero_data_to_add = np.zeros(shape=(len(dataset),len(dataset)-len(columns)))
        columns_to_add = ["temp_"+str(x) for x in range(0,len(dataset)-len(columns))]
        
        df_to_add = pd.DataFrame(zero_data_to_add, columns=columns_to_add,index=dataset.index,dtype='uint8')
        
        # add all prepared filler zeros to dataset
        dataset = pd.concat([dataset, df_to_add], axis=1)
        
        # shuffle rows and cols 
        np.random.seed(seed=13)
        items = dataset.columns
        
        random_column = np.random.permutation(dataset.shape[1])
        random_row = np.random.permutation(dataset.shape[0])
        
        print(f'\n\nRANDOM COLS: {random_column}\n\n')
        print(f'\n\nDataframe: {df.head()}\n\n')
        
        dataset.columns = [i for i in range(0,len(dataset.columns))]
        #print(dataset.columns)
        # recupero gli item selezionati nel relativo ordine == colonne
        items_reordered = [items[i] for i in random_column]
        # df selezionato e square
        # eliminare le righe nulle a priori
        df_square = dataset.iloc[random_row][random_column]
        # eliminare solo se items sensibili nulli
        
        f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
        f.set_figheight(8)
        f.set_figwidth(16)
        # check if sensitive items are null
        sensitive_list = df_square.columns[-num_sensitive:]
        
        # plot the sparse square matrix cut from the original dataset
        ax1.spy(df_square, marker='.', markersize='1')
        # ax1.show()
        
        # apply Reverse Cuthill-McKee (RCM) algorithm
        sparse = csr_matrix(df_square)
        order = reverse_cuthill_mckee(sparse)

        # reorder the items a second time after the RCM algorithm
        items_final = [items_reordered[i] for i in order]
        column_reordered = [df_square.columns[i] for i in order]
        items_final = dict(zip(column_reordered,items_final))

        # banded matrix
        df_square_band = df_square.iloc[order][column_reordered]
        
        # plot banded matrix 
        ax2.spy(df_square_band, marker='.', markersize='1')
        #ax2.show()
        plt.show()
        # initial matrix band
        [i, j] = np.where(df_square == 1)
        bw = max(i-j) + 1
        print("Bandwidth first RCM", bw)

        # matrix band after RCM
        [i, j] = np.where(df_square_band == 1)
        bw = max(i-j) + 1
        print("Bandwidth after RCM", bw)
        return df_square_band, items_final, sensitive_list
```

```python
def logger(label, obj):
    print(f'\n\n{"-"*20}\n{label}:\n{obj}\n{"-"*20}\n\n')
    
def symmetrize(a):
    """
    Return a symmetrized version of NumPy array a.

    Values 0 are replaced by the array value at the symmetric
    position (with respect to the diagonal), i.e. if a_ij = 0,
    then the returned array a' is such that a'_ij = a_ji.

    Diagonal values are left untouched.

    a -- square NumPy array, such that a_ij = 0 or a_ji = 0, 
    for i != j.
    """
    return a + a.T - np.diag(a.diagonal())
    
```
